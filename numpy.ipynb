```python
stack_arr1=np.array([1,2,3])
stack_arr2=np.array([2,3,5])
vstack_array=np.vstack((stack_arr1,stack_arr2))
print(vstack_array)
print(vstack_array.ndim)
hstack_array=np.hstack((stack_arr1,stack_arr2))
print(hstack_array)
```

```python
# boradcasting concepts
# if boradcasting is not used in python numpy manually use loop which are very slow for large data
prices=[100,200,300,400,500]
discount=10
final_prices=[]
for price in prices:
  after_discount=price-(price*discount/100)
  final_prices.append(after_discount)
print(final_prices)
# with broadcasting
price_array=np.array([100,200,300,400,500])
discount_array=10
final_price_array=price_array-(price_array*discount/100)
print(final_price_array)
```

```python
# vectorization
#without using numpy vectorization concepts
list1=[1,2,3]
list2=[4,5,6]
add_list=list1+list2# not add but manually add in a list [1,2,3,4,5,6]
print(add_list)
result=[x+y for x,y in zip(list1,list2)]
print(result)
# with numpy vectorization
add_arr1=np.array([1,2,3,4,5])
add_arr2=np.array([6,7,8,9,10])
result=add_arr1+add_arr2
print(result)
```
